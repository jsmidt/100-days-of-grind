```python
import gradio as gr
import numpy as np
import matplotlib.pyplot as plt



def create_plot(num_points):
    plt.ioff()
    x = np.linspace(0, 10, num_points)
    y = np.sin(x)
    fig, ax = plt.subplots()
    ax.plot(x, y)
    ax.set_title(f"Sine Wave ({num_points} points)")
    return fig

history = [
    {"role": "user", "content": "Please make me a plot of sin(x)."},
    {"role": "assistant", "content": "I am happy to provide you that report and plot."},
    {"role": "assistant", "content": gr.Plot(value=create_plot(100))},
    {"role": "assistant", "content": "Would you like anything else?"},
]

with gr.Blocks() as demo:
    gr.Chatbot(history, type="messages")

demo.launch()
```

```python
import gradio as gr

def load():
    return [
        ("Here's an audio", gr.Audio("https://github.com/gradio-app/gradio/raw/main/test/test_files/audio_sample.wav")),
        ("Here's an video", gr.Video("https://github.com/gradio-app/gradio/raw/main/demo/video_component/files/world.mp4"))
    ]

with gr.Blocks() as demo:
    chatbot = gr.Chatbot()
    button = gr.Button("Load audio and video")
    button.click(load, None, chatbot)

demo.launch()
```
